### Water Quality Prediction in India using Machine Learning Algorithms

```python
import numpy as np
import pandas as pd
```

```python
wq = pd.read_csv('Datasets/water_potability.csv')
```

```python
wq.info()
```

```python
wq.describe()
```

```python
ax = sns.countplot(x = "Potability",data= wq, saturation=0.8)
plt.xticks(ticks=[0, 1], labels = ["Not Potable", "Potable"])
plt.show()
```

```python
x = wq.Potability.value_counts()
labels = [0,1]
print(x)
```

```python
sns.violinplot(x='Potability', y='ph', data=wq, palette='bright')
```

```python
wq['ph'] = wq['ph'].fillna(wq['ph'].mean())
wq['Sulfate'] = wq['Sulfate'].fillna(wq['Sulfate'].mean())
wq['Trihalomethanes'] = wq['Trihalomethanes'].fillna(wq['Trihalomethanes'].mean())
```

```python
wq.head()
```

```python
wq.isnull().sum()
```

```python
from sklearn.svm import SVC, LinearSVC
```

```python
model_svm = SVC(kernel='rbf', random_state = 42)
```

```python
X = wq.drop('Potability', axis=1)
y = wq['Potability']
```

```python
X
```

```python
X.shape, y.shape
```

```python
# import StandardScaler to perform scaling
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
X
```

```python
# import train-test split 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

USING LOGISTICS REGRRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Creating model object
model_lg = LogisticRegression(max_iter=120,random_state=0, n_jobs=20)
```

```python
# Training Model
model_lg.fit(X_train, y_train)
```

```python
# Making Prediction
pred_lg = model_lg.predict(X_test)
```

```python
# Calculating Accuracy Score
lg = accuracy_score(y_test, pred_lg)
print(lg)
```

```python
print(classification_report(y_test,pred_lg))
```

```python
# confusion Maxtrix
cm1 = confusion_matrix(y_test, pred_lg)
sns.heatmap(cm1/np.sum(cm1), annot = True, fmt=  '0.2%', cmap = 'Reds')
```

USING DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Creating model object
model_dt = DecisionTreeClassifier( max_depth=4, random_state=42)
```

```python
# Training Model
model_dt.fit(X_train,y_train)
```

```python
# Making Prediction
pred_dt = model_dt.predict(X_test)
```

```python
# Calculating Accuracy Score
dt = accuracy_score(y_test, pred_dt)
print(dt)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
# confusion Maxtrix
cm2 = confusion_matrix(y_test, pred_dt)
sns.heatmap(cm2/np.sum(cm2), annot = True, fmt=  '0.2%', cmap = 'Reds')
```

USING RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Creating model object
model_rf = RandomForestClassifier(n_estimators=300,min_samples_leaf=0.16, random_state=42)
```

```python
# Training Model
model_rf.fit(X_train, y_train)
```

```python
# Making Prediction
pred_rf = model_rf.predict(X_test)
```

```python
# Calculating Accuracy Score
rf = accuracy_score(y_test, pred_rf)
print(rf)
```

```python
print(classification_report(y_test,pred_rf))
```

```python
# confusion Maxtrix
cm3 = confusion_matrix(y_test, pred_rf)
sns.heatmap(cm3/np.sum(cm3), annot = True, fmt=  '0.2%', cmap = 'Reds')
```

USING K NEAREST NEIGHBORS

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Creating model object
model_kn = KNeighborsClassifier(n_neighbors=9, leaf_size=20)
```

```python
# Training Model
model_kn.fit(X_train, y_train)
```

```python
# Making Prediction
pred_kn = model_kn.predict(X_test)
```

```python
# Calculating Accuracy Score
kn = accuracy_score(y_test, pred_kn)
print(kn)
```

```python
print(classification_report(y_test,pred_kn))
```

```python
# confusion Maxtrix
cm5 = confusion_matrix(y_test, pred_kn)
sns.heatmap(cm5/np.sum(cm5), annot = True, fmt=  '0.2%', cmap = 'Reds')
```
